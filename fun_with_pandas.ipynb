## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# Starter code...

%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")

#individual dataframes for the different data
df_phy_15 =pd.DataFrame(rawdata[0])
df_phy_10 =pd.DataFrame(rawdata[1])


# not important but df_cert = pd.DataFrame(rawdata[2])

df_dig_15 = pd.DataFrame(rawdata[3])
df_dig_10 = pd.DataFrame(rawdata[4])

#Combine the physical sales and digitalsales tables, and adds the marker for physical/digital:
df_phy = pd.concat([df_phy_15, df_phy_10])
df_phy['Format']='Physical'
df_dig = pd.concat([df_dig_15, df_dig_10])
df_dig['Format'] = 'Digital'

#Combine them both to one
df_complete_sales = pd.concat([df_phy, df_dig])

df_complete_sales
#df_complete_sales.astype('str').dtypes
#And because we don't need the sources...
#df_complete_sales.pop('Source')
```

```python
#Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
byear = 2001
byear_15 = byear +15

byear_entry = df_complete_sales[df_complete_sales['Released'] >= byear].sort_values(by='Released', ascending=True).iloc[0]

print(f"The bestselling release closest to my birthyear, {byear}, is the song {byear_entry['Single']} by {byear_entry['Artist']}, released in {byear_entry['Released']}.")
```

```python
#Outputs the artist and single name for the year you were 15 years old.
byear_15_entry = df_complete_sales[df_complete_sales['Released'] >= byear_15].sort_values(by='Released', ascending=True).iloc[0]
byear_15_entry
byear_15_entry['Artist']
print(f"The bestselling single when I was 15 years old was {byear_15_entry['Single']} by {byear_15_entry['Artist']}, released in released in {byear_15_entry['Released']}.")
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")

#Dataframe with all missions
df_space = pd.concat([pd.DataFrame(rawdata[0]),pd.DataFrame(rawdata[1]), pd.DataFrame(rawdata[2]), pd.DataFrame(rawdata[3]), pd.DataFrame(rawdata[4]), pd.DataFrame(rawdata[5]), pd.DataFrame(rawdata[6]), pd.DataFrame(rawdata[7]) ],ignore_index = True)
df_space
```

```python
#Seperating the dates and returning them to the dataframe
spacedate=list(df_space['Launch date'])
spacedate_seperated =[]
for date in spacedate:
    spacedate_seperated.append(date.split())
for date in spacedate_seperated:
    del date[0]

```

```python
#the splitting into months and years
space_month=[]
space_year=[]

for date in spacedate_seperated:
    space_month.append(date[0])
    space_year.append(date[1])

#now to add those two lists to the original dataframe, or perhaps make a new one
df_new_space = df_space
df_new_space.pop('Launch date')
df_new_space['Month']=space_month
df_new_space['Year']=space_year
```

```python
#Write a script that returns the year with the most launches
launch_count_year = df_new_space['Year'].value_counts()
launch_max_year = launch_count_year.idxmax()
year_count = launch_count_year.max()

print(f"The year with the most launches in {launch_max_year} with {year_count} launches")
```

```python
#Write a script that returns the most common month for launches
launch_count_month = df_new_space['Month'].value_counts()
launch_max_month = launch_count_month.idxmax()
month_count = launch_count_month.max()

print(f"The most common month for launches has been {launch_max_month} with {month_count} launches in total since the 1950's.")
```
